```python
ia.seed(1)

seq = iaa.Sequential(
    [
        iaa.Fliplr(0.5), # horizontal flip

        iaa.SomeOf((0, 2),
            [
                #iaa.PerspectiveTransform(scale=(0.003, 0.008)), # add perspective

                iaa.Affine( # zoom
                    scale=(1.0, 1.3),
                ),

                iaa.Sharpen(alpha=(0, 1.0), lightness=(0.90, 1.25)), # change alpha / lightness

                #iaa.OneOf([ # add blur / noise
                #    iaa.GaussianBlur((0.0, 1.8)),
                #    iaa.AdditiveGaussianNoise(
                #        loc=0, scale=(0.0, 0.02*255)),
                #]),

                #iaa.OneOf([ # change brightess / contrast / saturation
                #    iaa.AddToBrightness((-25, 25)),
                #    iaa.AllChannelsCLAHE(clip_limit=(1)),
                #    iaa.GammaContrast((0.75, 1.2)),
                #    iaa.MultiplySaturation((0.65, 1.3)),
                #]),

            ],
            random_order=True
        ),

        iaa.Sometimes(0.5, iaa.CoarseDropout((0.03, 0.06), size_percent=(0.01, 0.02))),
    ],
    random_order=False
)

seq2 = iaa.Sequential( # utilisé pour tester les augmentations individuellement
    [
        iaa.Sometimes(0.2, iaa.CoarseDropout((0.03, 0.06), size_percent=(0.07, 0.1))),
    ]
)
def py_augment(img, tgt):
    img = img.numpy()
    tgt = tgt.numpy()
    segmap = SegmentationMapsOnImage(tgt, shape=tgt.shape)
    img, tgt = seq(image=img, segmentation_maps=segmap)
    return img, tgt.get_arr()
```

### Préparation

```python
train_df = main_df[main_df["split"]=="train"]
val_df = main_df[main_df["split"]=="val"]
```

```python
steps_per_epoch = (train_df.shape[0] * data_passes) // (batch_size * n_epochs)

AUTOTUNE = tf.data.AUTOTUNE

BUFFER_SIZE = 1000

print(f"Training with {steps_per_epoch} steps per epoch")
```

```python
results_file = f'./results.csv'

if not os.path.isfile(results_file):
    results = pd.DataFrame(columns=['model_name', 'iou_score', 'time', 'loss'])
    results.to_csv(results_file, index=False)

def add_result(name, score, time, loss=None):
    result_df = pd.read_csv(results_file)
    result_df = result_df[result_df["model_name"] != name]
    result_df = result_df.append({"model_name": name, "iou_score": score, "time": time, "loss": loss}, ignore_index=True)
    result_df.to_csv(results_file, index=False)
    
def get_results():
    result_df = pd.read_csv(results_file)
    print(result_df.head(20))
    
def process_model(model, hist, name, time, safe_save=False):
    history = pd.DataFrame(data=hist.history)
    if not safe_save:
        model.save(f"./models/{name}.h5")
    model.save_weights(f'./models/{name}_weights.h5')
    history.to_csv(f'./models/{name}_history.csv', index=False)
    add_result(name,
               round(history.iloc[-1]["val_iou_score"], 4),
               time,
               round(history.iloc[-1]["val_loss"], 4))
    
def display_learning_curves(m_name, metric):
    hst = pd.read_csv(f'./models/{m_name}_history.csv')
    fig, ax = plt.subplots(1, 2, figsize=(20, 3))
    ax = ax.ravel()

    for i, met in enumerate([metric, 'loss']):
        ax[i].plot(hst[met])
        ax[i].plot(hst['val_' + met])
        ax[i].set_title('Model {}'.format(met))
        ax[i].set_xlabel('epochs')
        ax[i].set_ylabel(met)
        ax[i].legend(['train', 'val'])
    
    plt.show()
```

```python
def load_imgs(path1, path2, mode="original"):
    img1 = tf.io.read_file(path1)
    img1 = tf.io.decode_png(img1)
    img2 = tf.io.read_file(path2)
    img2 = tf.io.decode_png(img2)
    
    if mode == "crop" or mode == "augmented_crop":
        combined = tf.concat([img1, img2], axis=2)
        last_label_dim = tf.shape(img2)[-1]
        last_image_dim = tf.shape(img1)[-1]
        combined_crop = tf.image.random_crop(combined, size=tf.concat([CROP_SIZE, [last_label_dim + last_image_dim]], axis=0))
        img1 = combined_crop[:, :, :last_image_dim]
        img2 = combined_crop[:, :, last_image_dim:]
        
    if mode == "augmented_crop" or mode == "augmented_original":
        img1, img2 = tf.py_function(py_augment, inp=[img1, img2], Tout=[tf.uint8, tf.uint8])
        
    img1 = tf.cast(img1, tf.float32)
    img1 /= 255.
        
    img2 = tf.squeeze(img2, axis=-1)
    img2 = tf.one_hot(img2, 8, axis=-1)
    
    if mode == "original" or mode == "augmented_original":
        img1.set_shape((1024, 2048, 3))
        img2.set_shape((1024, 2048, 8))
    else:
        img1.set_shape((CROP_SIZE[1], CROP_SIZE[0], 3))
        img2.set_shape((CROP_SIZE[1], CROP_SIZE[0], 8))
    
    return img1, img2


def get_dataset(df, repeat=1, augment=0, original_size=0):
    ds = tf.data.Dataset.from_tensor_slices(
        (df["original_image_path"].values,
         df["labeled_image_path"].values,)
        ).shuffle(BUFFER_SIZE)

    if original_size:
        if augment:
            ds = ds.map(lambda x, y: load_imgs(x, y, "augmented_original"), num_parallel_calls=AUTOTUNE)
        else:
            ds = ds.map(lambda x, y: load_imgs(x, y, "original"), num_parallel_calls=AUTOTUNE)
    elif augment:
        ds = ds.map(lambda x, y: load_imgs(x, y, "augmented_crop"), num_parallel_calls=AUTOTUNE)
    else:
        ds = ds.map(lambda x, y: load_imgs(x, y, "crop"), num_parallel_calls=AUTOTUNE)
        
    if repeat:
        ds = ds.repeat()
        
    ds = ds.batch(batch_size)
    ds = ds.prefetch(buffer_size=AUTOTUNE)

    return ds
```

```python
def train_model(model, name, t_ds):
    
    val_ds = get_dataset(val_df, repeat=0)
    
    callbacks = [
             tf.keras.callbacks.ReduceLROnPlateau(monitor="iou_score", factor=0.2, patience=1, verbose=1, mode="max"),
             tf.keras.callbacks.EarlyStopping(monitor="iou_score", patience=3, mode="max", verbose=1, restore_best_weights=True)
    ]

    t0 = timer()

    hist = model.fit(t_ds,
                     epochs=n_epochs,
                     verbose=2,
                     validation_data=val_ds,
                     batch_size=batch_size,
                     steps_per_epoch=steps_per_epoch,
                     callbacks=callbacks)
    
    t1 = timer() - t0

    process_model(model, hist, name, t1)
```

```python
def prediction_display(mdl, loc=0):
    pred_ds = tf.data.Dataset.from_tensor_slices(
        ([val_df.iloc[loc]["original_image_path"]],
         [val_df.iloc[loc]["labeled_image_path"]],)
        ).shuffle(BUFFER_SIZE)
    pred_ds = pred_ds.map(load_imgs, num_parallel_calls=AUTOTUNE)
    pred_ds = pred_ds.batch(1)
    pred = mdl.predict(pred_ds)
    toimg = np.asarray(pred[0])
    img3 = np.argmax(toimg, axis=2)

    img1 = imageio.imread(val_df.iloc[loc]["original_image_path"])
    img2 = imageio.imread(val_df.iloc[loc]["labeled_image_path"])
    
    display_imgs(img1, img2, img3)
    
def cropped_prediction_display(mdl, loc=0):
    pred_ds = tf.data.Dataset.from_tensor_slices(
        ([val_df.iloc[loc]["original_image_path"]],
         [val_df.iloc[loc]["labeled_image_path"]],)
        ).shuffle(BUFFER_SIZE)
    pred_ds = pred_ds.map(lambda x, y: load_imgs(x, y, "augmented_crop"), num_parallel_calls=AUTOTUNE)
    pred_ds = pred_ds.batch(1)
    img, lbl = next(iter(pred_ds))
    pred = mdl.predict(img)
    img = img[0]
    lbl = lbl[0]
    pred = pred[0]
    img *= 255
    img = np.asarray(img, dtype=np.uint8)
    lbl = np.asarray(lbl)
    lbl = np.argmax(lbl, axis=2)
    lbl = np.expand_dims(lbl, axis=2)
    lbl = np.apply_along_axis(lambda x: color_map[x[0]] , -1, lbl)
    pred = np.asarray(pred)
    pred = np.argmax(pred, axis=2) 
    pred = np.expand_dims(pred, axis=2)
    pred = np.apply_along_axis(lambda x: color_map[x[0]] , -1, pred)
    plt.imshow(img)
    plt.show()
    plt.imshow(lbl)
    plt.show()
    plt.imshow(pred)
    plt.show()
```

### Baseline: Réseau convolutif avec encoder-decoder et saut de blocs

```python
def get_baseline(loss=sm.losses.bce_jaccard_loss):
    img_input = tf.keras.layers.Input(shape=(None, None, 3))
    
    # encoder
    
    conv1 = tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same')(img_input)
    conv1 = tf.keras.layers.Dropout(0.2)(conv1)
    conv1 = tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same')(conv1)
    pool1 = tf.keras.layers.MaxPooling2D((2, 2))(conv1)

    conv2 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(pool1)
    conv2 = tf.keras.layers.Dropout(0.2)(conv2)
    conv2 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(conv2)
    pool2 = tf.keras.layers.MaxPooling2D((2, 2))(conv2)
    
    # decoder
    
    conv3 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(pool2)
    conv3 = tf.keras.layers.Dropout(0.2)(conv3)
    conv3 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(conv3)
    up1 = tf.keras.layers.concatenate([tf.keras.layers.UpSampling2D((2, 2))(conv3), conv2], axis=-1)
    
    conv4 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(up1)
    conv4 = tf.keras.layers.Dropout(0.2)(conv4)
    conv4 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(conv4) 
    up2 = tf.keras.layers.concatenate([tf.keras.layers.UpSampling2D((2, 2))(conv4), conv1], axis=-1)
    
    # convolutions finales
    
    conv5 = tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same')(up2)
    conv5 = tf.keras.layers.Dropout(0.2)(conv5)
    conv5 = tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same')(conv5)
    
    # prediction
    
    out = tf.keras.layers.Conv2D(8, (1, 1) , padding='same', activation="softmax")(conv5)
    
    model = tf.keras.models.Model(inputs=img_input, outputs=out)
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=intitial_lr),
        loss=loss,
        metrics=[sm.metrics.iou_score],
    )
    
    return model
```

```python
if opti:
    train_dataset = get_dataset(train_df)
    val_dataset = get_dataset(val_df, repeat=0)
    
    param_grid = {
        "loss": [
            sm.losses.bce_dice_loss,
            sm.losses.bce_jaccard_loss,
            sm.losses.binary_focal_dice_loss,
            sm.losses.binary_focal_jaccard_loss,
            sm.losses.categorical_focal_dice_loss,
            sm.losses.categorical_focal_jaccard_loss
        ]
    }

    grid_search = GridSearch(
        model_fn=get_baseline,
        param_grid=param_grid,
        verbose=1,
        num_features=1,
        num_targets=1
    )

    grid_search.fit(
        train_dataset=train_dataset,
        val_dataset=val_dataset,
        epochs=1,
        verbose=1
    )

    grid_search.summary()
```
